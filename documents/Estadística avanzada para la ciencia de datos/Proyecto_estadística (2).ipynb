# 1.1 Calculation of financial variables and ratios

# Calculation of Net Income, Gross Profit, Revenue, COGS, EBIT

```python
# GROSS PROFIT = REVENUE - COGS
df_merge['Gross_Profit'] = df_merge['revenue'] - df_merge['cogs']

# EBIT = GROSS PROFIT - SGAE
df_merge['EBIT'] = df_merge['Gross_Profit'] - df_merge['sgae']

# NET INCOME = EBIT - FINEXP - INCOMETAX + OTHER INCOME + EXTRAORIDNARY ITEMS
df_merge['Net_Income'] = df_merge['EBIT'] - df_merge['finexp'] - df_merge['incometax'] + df_merge['otherincome'] + df_merge['extraordinaryitems']

# EBT
df_merge['EBT'] = df_merge['Gross_Profit'] - df_merge['sgae'] - df_merge['finexp'] + df_merge['otherincome'] + df_merge['extraordinaryitems']

df_merge[['Gross_Profit', 'EBIT', 'Net_Income']].head()
```

# Calculate market value and book value

```python
# MARKET VALUE = ORIGINALPRICE * SHARESOUTSTANDING
df_merge["marketvalue"] = df_merge["originalprice"] * df_merge["sharesoutstanding"]

# ORIGINALPRINCE = Precio original de la acción
# SHARESOUTSTANDING = # De acciones disponibles

# BOOK VALUE = TOTALASSETS - TOTALLIABILITIES
df_merge["bookvalue"] = df_merge["totalassets"] - df_merge["totalliabilities"]

df_merge.head()
```

Además, también calculamos los Retornos Continuos Compuestos. Estos se derivan utilizando logaritmos naturales para capturar los cambios porcentuales en los precios ajustados:

RETORNO TRIMESTRAL = QR = (ADJ PRICE / ADJ PRICE (T - 1) - 1)

RETORNO ANUAL = QR = (ADJ PRICE / ADJ PRICE (T - 4) - 1)

RETORNOS CONTINUOS COMPUESTOS:

RETURNS TRIMESTRE = QR = LOG(ADJ PRICE) - LOG(ADJ PRICE{T-1})

RETURNS ANUAL = QR = LOG(ADJ PRICE) - LOG(ADJ PRICE{T-4})

# ESCOGER 4 RATIOS DIFERENTES PARA METER AL MODELO DE REGRESIÓN

Return on Assets (ROA):

El ROA mide la eficiencia de una empresa para generar ganancias a partir de sus activos totales. Se calcula dividiendo el ingreso neto entre los activos totales. Este ratio es crucial para entender cuán bien una empresa está utilizando sus recursos para generar beneficios.

Gross Margin:

El Gross Margin es un indicador de rentabilidad que muestra el porcentaje de ingresos que excede el costo de los bienes vendidos (COGS). Este ratio se calcula dividiendo la utilidad bruta entre los ingresos, proporcionando una visión clara de la eficiencia con la que una empresa está manejando sus costos directos de producción.

Operating Profit Margin:

El Operating Profit Margin mide la rentabilidad de una empresa después de descontar los costos operativos, pero antes de considerar los intereses e impuestos. Se calcula dividiendo el EBIT (Earnings Before Interest and Taxes) entre los ingresos. Este ratio es esencial para evaluar la rentabilidad operativa de una empresa.

Interest Burden:

El Interest Burden ratio mide el impacto que los pagos de intereses tienen sobre las ganancias antes de impuestos (EBT). Se calcula dividiendo el EBT entre el EBIT. Un valor más bajo de este ratio sugiere que la empresa está soportando una carga de intereses significativa en comparación con sus ganancias operativas.

```python
# Return Ratios: Return on Assets (ROA)
df_merge['ROA'] = df_merge['Net_Income'] / df_merge['totalassets']
# Proftability Ratios: Gross Margin
df_merge['Gross_Margin'] = df_merge['Gross_Profit'] / df_merge['revenue']
# Profitability Ratios: Operating Profit Margin
df_merge['Operating_Profit_Margin'] = df_merge['EBIT'] / df_merge['revenue']
# Expense Ratios: Interest Burden
df_merge['Interest_Burden'] = df_merge['EBT'] / df_merge['EBIT']

print(df_merge[['ROA', 'Gross_Margin', 'Operating_Profit_Margin', 'Interest_Burden']].head())
```

```python
import numpy as np

# Using matrix algebra calculate the beta coefficients of your multiple regression model
X_matrix = np.array(X_clean)
y_vector = np.array(y_clean)

# Calcular (X^T X)^{-1} X^T y
beta = np.linalg.inv(X_matrix.T @ X_matrix) @ X_matrix.T @ y_vector
print("Coeficientes Beta:", beta)
```

```python
# Using matrix algebra calculate the standard errors of the beta coefficients
# Calcular los residuos
residuos = y_vector - X_matrix @ beta
sigma_squared = (residuos.T @ residuos) / (X_matrix.shape[0] - X_matrix.shape[1])

# Calcular la matriz de varianza-covarianza
var_covar_matrix = sigma_squared * np.linalg.inv(X_matrix.T @ X_matrix)

# Calcular errores estándar
standard_errors = np.sqrt(np.diag(var_covar_matrix))
print("Errores Estándar:", standard_errors)
```

La magnitud de mis coeficientes beta indica el efecto de cada variable  sobre la variable dependiente pero los errores estándar nos ayudan a determinar si estos efectos son estadísticamente significativos. Aquí estamos obteniendo resultados interesantes. Por un lado vemos el coeficiente beta de las variables Operating Profit Margin y EPS siendo los más altos en esta prueba. Curiosamente solo el de Operating Profit Margin es mucho mayor que su error estándar.  

```python
# Using matrix algebra calculate the R-squared and the adjusted R-squared of the model
# Calcular R-cuadrado
y_mean = np.mean(y_vector)
ss_total = np.sum((y_vector - y_mean) ** 2)
ss_residual = np.sum(residuos ** 2)
r_squared = 1 - (ss_residual / ss_total)
print("R-cuadrado:", r_squared)

# Calcular R-cuadrado ajustado
n = X_matrix.shape[0]
p = X_matrix.shape[1] - 1
r_squared_adjusted = 1 - ((1 - r_squared) * (n - 1) / (n - p - 1))
print("R-cuadrado Ajustado:", r_squared_adjusted)
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Supongamos que df_im y ratios ya están definidos como parte de tu código existente

# Definir las variables independientes (ratios financieros) y la variable dependiente (future_log_return)
X = df_im[ratios]
y = df_im['future_log_return']

# Añadir una constante al modelo (intercepto)
X_clean = sm.add_constant(X_clean)

# Método IQR para eliminar outliers
for column in ratios:
    Q1 = X_clean[column].quantile(0.25)
    Q3 = X_clean[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Eliminar filas que contienen outliers
    X_clean = X_clean[(X_clean[column] >= lower_bound) & (X_clean[column] <= upper_bound)]
    y_clean = y_clean[X_clean.index]

# Ajustar el modelo de regresión múltiple después de eliminar outliers
model = sm.OLS(y_clean, X_clean).fit()
print("Modelo después de eliminar outliers con IQR")
print(model.summary())
```

El uso del método IQR fue mi primera aproximación para mitigar el problema de los outliers. Esto fue necesario debido a la presencia de valores atípicos en variables como Operating Profit Margin y EPS, detectados en los análisis anteriores como en la sección de Media Aritmética, Mediana  y Promedio Ponderado. A diferencia de la winsorización, este método es útil cuando queremos mantener la mayor cantidad de datos posibles sin alterar significativamente la estructura de los datos. Mi expectativa con este método es que elimine los valores extremos sin comprometer el modelo general y quiero que R2 aumente.



El modelo ajustado tras aplicar el método IQR mostró un R-cuadrado de 0.008, una ligera mejora comparado con los modelos anteriores, lo que sugiere una mejora marginal en la capacidad del modelo para explicar la variabilidad del future_log_return. Algunas variables, como ROA (coeficiente: 0.4489, p-valor: 0.000) y Gross Margin (coeficiente: 0.0337, p-valor: 0.000), ganaron significancia estadística después de la eliminación de los outliers, lo que indica que, al mitigar la influencia de los valores extremos, estas relaciones se volvieron más claras. Sin embargo, otras variables como EPS (p-valor: 0.556) siguen sin ser significativas. En general, el método IQR parece haber limpiado el modelo de valores distorsionantes, mejorando la interpretación de ciertas variables clave, sin embargo, procederé a intentar winsorización.

```python
from scipy.stats import mstats

# Winsorización para el 5% superior e inferior
X_winsorized = X_clean.apply(lambda x: mstats.winsorize(x, limits=[0.1, 0.1]), axis=0)

# Reajustar el modelo de regresión después de la winsorización
model_winsorized = sm.OLS(y_clean, X_winsorized).fit()

# Mostrar el resumen del nuevo modelo
print(model_winsorized.summary())
```

El siguiente paso fue aplicar la winsorización con un límite del 10% para evaluar cómo el modelo responde cuando se ajustan los valores extremos en todas las variables. A diferencia del método IQR, aquí los valores atípicos no se eliminan, sino que se limitan al rango de percentiles definido. Esto nos permite mitigar la influencia de los outliers sin perder datos importantes. La winsorización suele ser útil cuando los outliers son extremos pero no necesariamente incorrectos o erróneos como puede ser el caso debido al contexto de las empresas de USA.

Después de aplicar la winsorización, observamos una ligera mejora en el R-cuadrado, que alcanzó 0.009, lo que sugiere que la capacidad explicativa del modelo mejoró un poco más respecto al método IQR. Varias variables clave se volvieron estadísticamente significativas, como Operating Profit Margin (coeficiente: 0.0962, p-valor: 0.000) y Interest Burden (coeficiente: -0.0476, p-valor: 0.000). ***CORRECCIÓN***

se optó por realizar una imputación de valores faltantes mediante la mediana de cada empresa (agrupada por la variable firm), en lugar de imputar los valores faltantes a nivel general para toda la muestra o simplemente eliminar los valores.

Justificación

Las empresas en el conjunto de datos pertenecen a diferentes sectores y tienen estructuras financieras únicas. La imputación general sobre toda la muestra podría introducir sesgos, ya que los valores de empresas más grandes o más rentables podrían distorsionar los resultados. Al agrupar por empresa, se asegura que la imputación refleje las características específicas de cada entidad.

La mediana es menos sensible a los valores atípicos en comparación con la media. Dado que algunas empresas pueden tener trimestres o años atípicos, este enfoque asegura que los valores atípicos no afecten de manera desproporcionada el proceso de imputación.

En el caso de empresas con periodos consecutivos con valores faltantes, imputar por la mediana de la misma empresa ayuda a mantener la consistencia en los patrones financieros y operativos de la misma. Esto es especialmente importante en los cálculos de ratios como el margen bruto o el retorno sobre activos, que varían considerablemente de empresa a empresa.

La imputación por mediana dentro de cada grupo de empresas asegura que los datos imputados estén más alineados con los valores reales esperados para cada entidad, lo que puede mejorar la precisión y robustez del modelo de regresión al evitar introducir información de otras empresas que no refleja la realidad financiera de cada una.

```python
import statsmodels.api as sm
import numpy as np

# Imputar valores NaN con la mediana de cada empresa (firm)
X_imputed = X.copy()
X_imputed['firm'] = df_filtered['firm']

# Aplicar la imputación por cada grupo de empresa
X_imputed = X_imputed.groupby('firm').transform(lambda x: x.fillna(x.median()))

# Reemplazar inf y -inf por NaN en X y y para eliminarlos después
X_imputed = X_imputed.replace([np.inf, -np.inf], np.nan)
y_clean = y.replace([np.inf, -np.inf], np.nan)

# Mantener solo las filas que no tienen NaN o inf en X e y
X_clean = X_imputed.loc[y_clean.dropna().index].dropna()
y_clean = y_clean.loc[X_clean.index]

# Agregar una constante para el modelo
X_clean = sm.add_constant(X_clean)

# Ajustar el modelo de regresión con los datos limpios
model_clean = sm.OLS(y_clean, X_clean).fit()

# Mostrar el resumen del modelo
print(model_clean.summary())
```

```python
import pandas as pd

df = pd.read_csv("conjunto_de_datos_igae_igae2024_07.csv")
df.head()
```

```python
import pandas as pd

# Filtrar la fila exacta que contiene el descriptor "Índice de volumen físico base 2018=100|Total"
df_filtered = df[df['Descriptores'] == "Índice de volumen físico base 2018=100|Total"]

# Transponer los datos para obtener una serie temporal
df_melted = df_filtered.melt(id_vars=['Descriptores'], var_name='Fecha', value_name='IGAE')

# Separar la columna 'Fecha' en dos: Año y Mes
df_melted[['Año', 'Mes']] = df_melted['Fecha'].str.split('|', expand=True)

# Diccionario para mapear meses en español a números
meses_map = {
    'Enero': '01', 'Febrero': '02', 'Marzo': '03', 'Abril': '04',
    'Mayo': '05', 'Junio': '06', 'Julio': '07', 'Agosto': '08',
    'Septiembre': '09', 'Octubre': '10', 'Noviembre': '11', 'Diciembre': '12'
}

# Reemplazar el nombre del mes por su valor numérico
df_melted['Mes'] = df_melted['Mes'].map(meses_map)

# Crear una nueva columna de fecha en formato 'YYYY-MM'
df_melted['Fecha'] = df_melted['Año'] + '-' + df_melted['Mes']

# Convertir la columna 'Fecha' a formato datetime
df_melted['Fecha'] = pd.to_datetime(df_melted['Fecha'], format='%Y-%m')

# Eliminar las columnas auxiliares de Año y Mes si ya no son necesarias
df_melted = df_melted.drop(columns=['Año', 'Mes'])

# Revisar el DataFrame final
df_melted.head()
```

# Testing for stationary

```python
import numpy as np

# Aplicar el logaritmo natural a la serie IGAE
df_melted['Log_IGAE'] = np.log(df_melted['IGAE'])

# Graficar el resultado del logaritmo natural
plt.figure(figsize=(10,6))
plt.plot(df_melted.index, df_melted['Log_IGAE'], label='Logaritmo Natural de IGAE')
plt.title('Logaritmo Natural de IGAE Over Time')
plt.xlabel('Fecha')
plt.ylabel('Log(IGAE)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Aplicar diferenciación estacional (lag=12) al logaritmo natural
df_melted['Diff_Log_IGAE'] = df_melted['Log_IGAE'].diff(12)

# Graficar la diferenciación estacional
plt.figure(figsize=(10,6))
plt.plot(df_melted.index, df_melted['Diff_Log_IGAE'], label='Diferenciación Estacional (lag=12)')
plt.title('Diferenciación Estacional del Logaritmo de IGAE')
plt.xlabel('Fecha')
plt.ylabel('Diferenciación Estacional')
plt.legend()
plt.grid(True)
plt.show()
```

ESTA GRÁFICA REPRESENTA EL CAMBIO PORCENTUAL DEL LOG IGAE CADA AÑO

```python
from statsmodels.tsa.stattools import adfuller

# Aplicar la prueba de Dickey-Fuller en la serie diferenciada (Diff_Log_IGAE)
adf_test = adfuller(df_melted['Diff_Log_IGAE'].dropna())

# Mostrar los resultados
print('ADF Statistic:', adf_test[0])
print('p-value:', adf_test[1])
for key, value in adf_test[4].items():
    print(f'Critical Value {key}: {value}')
```

Interpretación:
El p-valor es extremadamente bajo (mucho menor que 0.05), lo que nos permite rechazar la hipótesis nula de que la serie es estacionaria.


# ACF and PACF plots

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Graficar la ACF (Autocorrelation Function)
plt.figure(figsize=(10,6))
plot_acf(df_melted['Diff_Log_IGAE'].dropna(), lags=48, ax=plt.gca())
plt.title('ACF - Autocorrelation Function')
plt.show()

# Graficar la PACF (Partial Autocorrelation Function)
plt.figure(figsize=(10,6))
plot_pacf(df_melted['Diff_Log_IGAE'].dropna(), lags=60, ax=plt.gca())
plt.title('PACF - Partial Autocorrelation Function')
plt.show()
```

ACF (Autocorrelation Function):
La ACF mide la correlación entre una observación y los valores pasados de la serie a diferentes rezagos (lags). Las primeras 5 autocorrelaciones son significativas, lo que indica que los valores de la serie IGAE están correlacionados con los 5 rezagos anteriores. Esto significa que la serie actual tiene una fuerte correlación con los valores de los últimos 5 meses.
El primer rezago tiene una autocorrelación de casi 0.8, lo que sugiere que la diferencia estacional de IGAE está fuertemente correlacionada con el valor del mes anterior. El decaimiento gradual de las autocorrelaciones después del rezago 5 podría ser indicativo de un componente MA (Moving Average) en el modelo. Esto sugiere que incluir un término de media móvil (MA) en el modelo podría capturar este comportamiento de la serie.

PACF (Partial Autocorrelation Function):
La PACF mide la correlación parcial, es decir, la relación directa entre una observación y los valores anteriores, excluyendo los efectos intermedios de otras observaciones. El primer rezago es muy significativo (aproximadamente 0.8), lo que indica una fuerte correlación entre la diferencia estacional de IGAE y su primer rezago. Esto sugiere la presencia de un componente AR (Autoregressive) en el modelo. Los rezagos posteriores son menores, pero los primeros 2 rezagos tienen cierta significancia. Esto sugiere que es posible un modelo con un componente autoregresivo de orden 1 o 2.

# Running the ARIMA/SARIMA

```python
# Ejemplo: Crear una lista con las fechas de crisis (modifica según sea necesario)
crisis_months = ['1995-01', '1995-02', '2008-09', '2008-10', '2020-03', '2020-04']  # Fechas de crisis conocidas
# 1995-01 y 1995-02: Crisis del "Efecto Tequila" o crisis del peso mexicano
# 2008-09 y 2008-10: Crisis financiera global
# 2020-03 y 2020-04: Impacto de la pandemia de COVID-19

# Crear la columna de variable dummy de crisis
df_melted['crisis_dummy'] = df_melted['Fecha'].apply(lambda x: 1 if str(x)[:7] in crisis_months else 0)

# Verificar los primeros datos
print(df_melted[['Fecha', 'crisis_dummy']].head())
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Definir el modelo SARIMA con la variable exógena (dummy de crisis)
model = SARIMAX(df_melted['Diff_Log_IGAE'],
                exog=df_melted[['crisis_dummy']],  # Aquí se agrega la variable exógena
                order=(2, 1, 0),               # (p, d, q)
                seasonal_order=(1, 0, 0, 12),  # (P, D, Q, S)
                enforce_stationarity=False,
                enforce_invertibility=False,
                trend='c')  # 'c' para constante

# Ajustar el modelo a los datos
results = model.fit()

# Resumen del modelo
print(results.summary())

# CORRECCIÓN
# AGREGUÉ LA CONSTANTE
# HICE EL MODELO CON LA VARIABLE DE CAMBIO PORCENTUAL COMO SE DEBÍA
# AGREGUÉ P 1
```

d = 1: La diferenciación se aplicó una vez para eliminar cualquier tendencia a largo plazo y hacer la serie estacionaria. La prueba de Dickey-Fuller mostró que la serie diferenciada era estacionaria, por lo que una diferenciación fue suficiente.

q = 0: En la ACF no había correlaciones significativas a largo plazo, por lo que no incluimos un término de media móvil (MA).
